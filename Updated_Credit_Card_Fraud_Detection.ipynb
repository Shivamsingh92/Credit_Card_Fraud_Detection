Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

Splitting the data into Features & Targets

```python
X = credit_card_data.drop(columns='Class', axis=1)
Y = credit_card_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.70,  random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

# Under Sampling

```python
# Implementing Undersampling for Handling Imbalanced
from imblearn.under_sampling import NearMiss
nm = NearMiss()
X_train_us,y_train_us=nm.fit_resample(X_train,y_train)
from collections import Counter
print("The number of classes before fit{}".format(Counter(y_train)))
print("The number of classes after fit{}".format(Counter(y_train_us)))
```

#Model Evaluation

Logistics Regression

```python
model = LogisticRegression()
model.fit(X_train_us, y_train_us)
X_test_prediction = model.predict(X_test)
print(classification_report(y_test,X_test_prediction))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train_us, y_train_us)
X_test_prediction = model.predict(X_test)
print(classification_report(y_test,X_test_prediction))
```

SMOTE

```python
# Implementing Undersampling for Handling Imbalanced
from imblearn.over_sampling import SMOTE
sm = SMOTE()
X_train_us,y_train_us=sm.fit_resample(X_train,y_train)
from collections import Counter
print("The number of classes before fit{}".format(Counter(y_train)))
print("The number of classes after fit{}".format(Counter(y_train_us)))
```

Model Evaluation

```python
model = LogisticRegression()
model.fit(X_train_us, y_train_us)
X_test_prediction = model.predict(X_test)
print(classification_report(y_test,X_test_prediction))
```
